```python
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
```

# GPS Measurements

```python
#Get gps consumption (Joule) with a batch technique
gps_batch = []
with open('gps_batch.txt', 'r') as file:
    for line in file:
        gps_batch.extend(line.strip().replace(',', '.').split(';'))

gps_batch = [float(x) for x in gps_batch if x] 

print("Number of measurements: " + str(len(gps_batch)))
print("Total consumption : " + str(sum(gps_batch)))
```

```python
#Get gps consumption (Joule) without batch technique
gps_no_batch = []
with open('gps_no_batch.txt', 'r') as file:
    for line in file:
        gps_no_batch.extend(line.strip().replace(',', '.').split(';'))

gps_no_batch = [float(x) for x in gps_no_batch if x] 

print("Number of measurements: " + str(len(gps_no_batch)))
print("Total consumption : " + str(sum(gps_no_batch)))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Histogram 1
axs[0].hist(gps_batch, bins=30, color='blue', edgecolor='black')
axs[0].set_title('Distribution of GPS data (Batch)')
axs[0].set_xlabel('Value')
axs[0].set_yscale('log')
axs[0].set_ylabel('Frequency')

# Histogram 2
axs[1].hist(gps_no_batch, bins=30, color='blue', edgecolor='black')
axs[1].set_title('Distribution of GPS data (Without Batch)')
axs[1].set_xlabel('Value')
axs[1].set_yscale('log')
axs[1].set_ylabel('Frequency')

# Layout Optimization
plt.tight_layout()
plt.show()
```

# Mann–Whitney U Test

Mann–Whitney U test (also called the Mann–Whitney–Wilcoxon, Wilcoxon rank-sum test, or Wilcoxon–Mann–Whitney test) is a nonparametric test of the null hypothesis that, for randomly selected values X and Y from two populations, the probability of X being greater than Y is equal to the probability of Y being greater than X.

We use U-Test instead of T-Test becouse data isn't Normal Distributed.

```python
#For U-Test we need the same samples size for gps_batch and gps_no_batch
if len(gps_batch) > len(gps_no_batch):
    gps_batch = gps_batch[0:len(gps_no_batch)]
else: 
    gps_no_batch = gps_no_batch[0:len(gps_batch)]

print("Size of samples: " + str(len(gps_batch)) + "\n")

print("Null Hypotesis: method with batch and without batch have the same energy impact")
print("Alternative Hypotesis: method with batch and without batch have different energy impact \n")
    
group_1 = gps_batch
group_2 = gps_no_batch

# Make a U-Test of Mann-Whitney
statistic_U, p_value = mannwhitneyu(group_1, group_2)

print("Statistic U of Mann-Whitney:", statistic_U)
print(f"P-Value: {p_value:.6}")
print("")

# Set significance value alpha = 0.01
if p_value < 0.01:
    print("Significant Statistical Difference, we reject the null hypotesis")
else:
    print("Non-Significant Statistical Difference, we accept the null hypotesis")
```
